

## üåê What is an HTML Form?

* An HTML form collects **user input** (text, password, email, etc.) and sends it to the server.
* Key attributes of `<form>`:

  1. **method** ‚Üí HTTP method: `GET` or `POST`
  2. **action** ‚Üí URL where the form data is sent (optional; defaults to current URL)

---

## ‚úÖ 1Ô∏è‚É£ Basic Form Structure

```html
<form method="POST" action="/submit">
    <label for="username">Username:</label>
    <input type="text" id="username" name="username" required><br><br>

    <label for="password">Password:</label>
    <input type="password" id="password" name="password" required><br><br>

    <input type="submit" value="Submit">
</form>
```

**Explanation:**

* `name` ‚Üí key used by Flask to access the value (`request.form["username"]`)
* `required` ‚Üí makes the field mandatory
* `type="submit"` ‚Üí creates a submit button

---

## ‚úÖ 2Ô∏è‚É£ Flask Route to Handle Form

**app.py**

```python
from flask import Flask, request, render_template

app = Flask(__name__)

@app.route("/login", methods=["GET", "POST"])
def login():
    if request.method == "POST":
        username = request.form["username"]
        password = request.form["password"]
        return f"<h1>Welcome, {username}!</h1><p>Your password is {password}</p>"
    return render_template("login.html")

if __name__ == "__main__":
    app.run(debug=True)
```

* **GET request** ‚Üí shows the form
* **POST request** ‚Üí handles submitted data

## ‚úÖ 3Ô∏è‚É£ Full Example with Template

**templates/login.html**

```html
<!DOCTYPE html>
<html>
<head>
    <title>Login Form</title>
</head>
<body>
    <h1>Login</h1>
    <form method="POST">
        <label>Username:</label>
        <input type="text" name="username" required><br><br>

        <label>Password:</label>
        <input type="password" name="password" required><br><br>

        <input type="submit" value="Login">
    </form>
</body>
</html>
```

**How it works:**

1. User enters **username** and **password**
2. Clicks **Submit**
3. Flask receives the data via `request.form`

## ‚úÖ 4Ô∏è‚É£ GET vs POST Forms

| Feature         | GET                   | POST                 |
| --------------- | --------------------- | -------------------- |
| Data in URL     | Yes (visible)         | No (hidden)          |
| Use case        | Search forms, filters | Login, registration  |
| Security        | Low (visible data)    | Higher (hidden data) |
| Browser caching | Yes                   | No                   |

**Example GET form:**

```html
<form method="GET" action="/greet">
    <input type="text" name="name">
    <input type="submit" value="Say Hello">
</form>
```

* Flask retrieves: `request.args.get("name")`

---

### üîé Key Points

1. **Every input field must have a `name`** for Flask to access it.
2. **Submit button** sends form data to the server.
3. Combine forms with **variables, loops, and conditionals** in templates to make interactive pages.
