# FLASK


## üåê What is Flask?

* **Flask** is a **lightweight web framework** in Python used to build web applications (websites, APIs, dashboards, etc.).
* It is called a **micro-framework** because it doesn‚Äôt come with a lot of built-in tools like authentication, database layers, or form validation ‚Äî instead, you can add only what you need.
* Created by **Armin Ronacher** in 2010.

Think of Flask as a **toolbox for creating websites** with Python.

---

## ‚úÖ Key Features of Flask

* **Lightweight and flexible** ‚Äì You can start small and grow big.
* **Built-in web server** ‚Äì No need to install extra servers for development.
* **Routing** ‚Äì Easy to define URL routes (`/home`, `/about`, etc.).
* **Jinja2 templating** ‚Äì Helps embed Python logic inside HTML pages.
* **Extensible** ‚Äì Can integrate with databases (MySQL, MongoDB, SQLite), authentication libraries, etc.
* **Good for APIs** ‚Äì Flask is widely used to create RESTful APIs.

## üöÄ A Simple Flask Example

Here‚Äôs the most basic Flask app:

```python
from flask import Flask

app = Flask(__name__)   # create a Flask app

@app.route("/")         # define a route (homepage)
def home():
    return "Hello, Flask!"   # response

if __name__ == "__main__":
    app.run(debug=True)   # run the app in debug mode
```

### How it works:

1. `Flask(__name__)` ‚Üí Creates the web application.
2. `@app.route("/")` ‚Üí Defines the URL route (`/` means homepage).
3. The `home()` function returns **text (or HTML)** to show on the page.
4. `app.run(debug=True)` ‚Üí Starts a local server at **[http://127.0.0.1:5000/](http://127.0.0.1:5000/)**.

## üìÑ Adding HTML Templates

Flask supports **templates** (HTML with Python variables).

Folder structure:

```
project/
  app.py
  templates/
    index.html
```

**app.py**

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/")
def home():
    return render_template("index.html", name="Revathy")
```

**templates/index.html**

```html
<!DOCTYPE html>
<html>
<head>
    <title>My Flask App</title>
</head>
<body>
    <h1>Hello, {{ name }}!</h1>
</body>
</html>
```

When you open your app, it will show:

```
Hello, Revathy!
```
