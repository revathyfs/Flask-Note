

## üåê URL Parameters in Flask

In Flask, you can define a route like:

```python
@app.route("/user/<name>")
```

Here, `<name>` is a **dynamic part of the URL**, and Flask passes it as an argument to the view function.

---

### ‚úÖ Example Code

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "<h1>Welcome!</h1><p>Go to /user/YourName</p>"

# Dynamic URL parameter
@app.route("/user/<name>")
def user_profile(name):
    return f"<h1>Hello, {name}!</h1><p>Welcome to your profile page.</p>"

if __name__ == "__main__":
    app.run(debug=True)
```

---

### üîé How it works

* Visit: `http://127.0.0.1:5000/user/Revathy`
  Output:

  ```
  Hello, Revathy!
  Welcome to your profile page.
  ```
* Flask automatically **extracts the part after `/user/`** and passes it as the `name` argument to `user_profile()`.

---

### ‚úÖ Optional: Type-Specific Parameters

You can enforce types like `int` or `float`:

```python
@app.route("/square/<int:number>")
def square(number):
    return f"The square of {number} is {number**2}"
```

* Visit: `/square/5` ‚Üí Output: `The square of 5 is 25`
* If you try `/square/abc` ‚Üí 404 error (because it expects an integer)

Supported converters:

* `string` (default) ‚Üí any text
* `int` ‚Üí integer
* `float` ‚Üí decimal numbers
* `path` ‚Üí allows slashes `/`

---

### ‚úÖ Using URL Parameters with Templates

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/user/<name>")
def user_profile(name):
    return render_template("user.html", username=name)
```

**templates/user.html**

```html
<h1>Hello, {{ username }}!</h1>
<p>Welcome to your profile page.</p>
```

---

‚úÖ Result: You now have **dynamic pages** where each URL can show different content based on the parameter.
